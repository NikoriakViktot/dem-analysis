
# Опис файлів TanDEM-X

Файли, про які йдеться, є частинами продукту **TanDEM-X**, який надає відредаговані дані висот за допомогою супутників **TanDEM-X** та **TerraSAR-X**. Ось роз'яснення стосовно двох файлів:

## 1. **TDM1_EDEM_10_N47E024_EDEM_W84.tif**
   - **Опис**: Це відредаговані дані DEM (цифрова модель рельєфу), що представлені у висотах відносно **WGS84**, тобто у **еліпсоїдних висотах**.
   - **Призначення**: Використовується для візуалізації даних на основі еліпсоїдної системи координат (WGS84), яка є стандартною геодезичною системою.
   - **Формат**: GeoTIFF, тип даних — **FLOAT**, з можливістю роботи з великими обсягами даних у **Cloud-Optimized GeoTIFF** форматі.

## 2. **TDM1_EDEM_10_N47E024_EDEM_EGM.tif**
   - **Опис**: Цей файл містить відредаговані дані DEM, але в цьому випадку висоти приведені до **геоїдної системи EGM96** (Earth Gravitational Model 1996).
   - **Призначення**: Висоти у цьому файлі приведені до геоїда, що є корисним для точнішого геодезичного аналізу, оскільки геоїд наближається до середнього рівня моря, і такі дані корисні для багатьох застосунків, включаючи гідрологічні та океанографічні дослідження.
   - **Формат**: Також GeoTIFF, з типом даних **FLOAT**.

## Різниця між файлами:
- **Еліпсоїдні висоти (WGS84)**: Використовуються для загальних геодезичних і картографічних задач, не враховуючи вплив гравітації.
- **Геоїдні висоти (EGM96)**: Цей файл враховує гравітаційний вплив Землі, тому його точність є більшою для задач, де важлива прив'язка до рівня моря.

## Пояснення до XML метаданих:
- У метаданих, які ти навів, **TDM1_EDEM_10_N47E024_EDEM_W84.tif** і **TDM1_EDEM_10_N47E024_EDEM_EGM.tif** — це два різні формати висот, один для **еліпсоїдних висот**, а інший — для **геоїдних висот**.
- Окрім цього, є також файли для **Height Error Map (HEM)** та **Editing Mask (EDM)**, які допомагають зрозуміти, де в DEM є помилки або де було редагування.

## Підсумок:
- **EDEM_W84** — це версія DEM, де висоти відображаються відносно стандартного еліпсоїда WGS84.
- **EDEM_EGM** — це версія DEM, де висоти приведені до геоїда EGM96, що точніше відображає рівень моря для деяких застосувань.

Обидва файли можна використовувати для різних типів аналізу, залежно від того, яка система координат вам потрібна для конкретної задачі.
